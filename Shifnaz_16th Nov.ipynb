## Q17. List the most frequent routes(same START & END points) taken by Uber drivers. 
## (3 points)

```python
a = uberdata.loc[(uberdata['START*'] == uberdata['STOP*'])]
b = a.groupby(by='START*').count()
frequent_routes = b.sort_values(by='START_DATE*', ascending=False).head()
frequent_routes['START_DATE*']
```

## Q22. What is proportion of trips that is Business and what is the proportion of trips that is Personal? (3 points)

```python
total = uberdata['CATEGORY*'].count() 
business = uberdata[uberdata['CATEGORY*']=='Business'].count()
proportion_of_business = (business/total)*100
proportion_of_business['CATEGORY*']
```

```python
total = uberdata['CATEGORY*'].count() 
personal = uberdata[uberdata['CATEGORY*']=='Personal'].count()
proportion_of_personal = (personal/total)*100
proportion_of_personal['CATEGORY*']
```
